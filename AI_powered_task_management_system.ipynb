```python
def normalize_nested_column(df, column_name):
    # Explode the column if it contains lists
    df = df.explode(column_name)

    # If the column contains dictionaries, normalize them
    if df[column_name].apply(lambda x: isinstance(x, dict)).any():
        normalized = pd.json_normalize(df[column_name])
        normalized.columns = [f"{column_name}_{subcol}" for subcol in normalized.columns]
        df = df.drop(columns=[column_name]).reset_index(drop=True)
        df = pd.concat([df, normalized], axis=1)
    else:
        # If the column contains scalar values, just rename it
        df = df.rename(columns = {column_name: f"{column_name}_value"})

    return df

for column in columns_with_nested_data:
    df = normalize_nested_column(df, column)
```

```python
import nltk
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.stem import WordNetLemmatizer



nltk.download("stopwords")
nltk.download("punkt")
nltk.download("wordnet")  # Required for lemmatization

lemmatizer = WordNetLemmatizer()
tokenizer = TweetTokenizer()
```

```python
stop_words = set(stopwords.words("english"))
stop_words
```

```python
def clean_text(text):

    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", " ", text, flags=re.MULTILINE)  # Remove URLs

    tokenizer = TweetTokenizer()
    words = tokenizer.tokenize(text)
    words = [word for word in words if word not in stop_words]  # Remove stopwords

    text = " ".join(words)  # Reconstruct text after stopword removal
    text = re.sub(r"[^\w\s]", "", text)  # Removes punctuation
    text = re.sub(r"\d+", " ", text)  # Removes all digits
    text = text.replace("_", " ")

    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word, wordnet.VERB) for word in text.split()]  # Lemmatize

    return " ".join(lemmatized_words)


df["clean_desc"] = df["desc"].apply(clean_text)
```

```python
vectorizer = TfidfVectorizer(max_features = 500)
X_tfidf = vectorizer.fit_transform(df["clean_desc"])

# Convert the TF-IDF matrix to a DataFrame for better visualization
df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
df_tfidf
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

X = df_tfidf  # Features
y = df["labels_name"]  # Target

# Apply SMOTE
smote = SMOTE(sampling_strategy='auto', k_neighbors=2, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("Balanced class distribution:", Counter(y_resampled))
```

```python
from sklearn.model_selection import train_test_split

# Split data into Training (80%) & Testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.2, random_state = 42)
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

# Initialize SVM with a linear kernel
svm = SVC(kernel = "linear",probability=True)

# Apply RFE to select top features
rfe = RFE(estimator = svm, n_features_to_select = 100)
X_selected = rfe.fit_transform(X_train, y_train)
rfe = rfe.get_support(indices = True)
rfe = X_train.columns[rfe]

print("Selected features:", rfe)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

X_train = X_train[rfe]
X_test = X_test[rfe]

# NaÃ¯ve Bayes Tuning
nb_params = {"alpha": [0.001, 0.01, 0.1, 0.5, 1, 10]}
nb_grid = GridSearchCV(MultinomialNB(), nb_params, cv = 5, scoring = "accuracy")
nb_grid.fit(X_train, y_train)
nb_best = nb_grid.best_estimator_
print(nb_best, "\n")

# SVM Tuning
svm_params = {"C": [0.001, 0.01, 0.1, 1, 10], "gamma": [0.001, 0.01, 0.1, 1, 10], "kernel": ["linear", "rbf"]}
svm_grid = GridSearchCV(SVC(), svm_params, cv = 5, scoring = "accuracy")
svm_grid.fit(X_train, y_train)
svm_best = svm_grid.best_estimator_
print(svm_best, "\n")

nb_model = nb_best
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)

print("NaÃ¯ve Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
print("\nClassification Report:\n", classification_report(y_test, y_pred_nb))

svm_model = svm_best
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("\nClassification Report:\n", classification_report(y_test, y_pred_svm))
```

```python
def predict_task_attributes(task_description):

    clean_desc = clean_text(task_description)

    tfidf_vector = vectorizer.transform([clean_desc])

    tfidf_df = pd.DataFrame(tfidf_vector.toarray(), columns=vectorizer.get_feature_names_out())

    X_processed = tfidf_df[rfe]

    svm_prediction = svm_model.predict(X_processed)[0]

    labels_name_to_priority_heuristic = {
        "Power-Ups": 1,
        "REST API": 2,
        "Update": 3,
        "Addition": 4,
        "Breaking Change": 3,
        "Deprecation": 2,
        "Deletion": 1,
        "Bug": 5
    }

    predicted_priority_svm = labels_name_to_priority_heuristic.get(svm_prediction, "Unknown")


    return {
        "task_description": task_description,
        "clean_description": clean_desc,
        "svm_classification": svm_prediction,
        "svm_priority": predicted_priority_svm
    }
```

```python
df_priority = df_tfidf.copy()
df_priority[["badges_attachments", "pos"]] = df[["badges_attachments", "pos"]]

df_priority["year"] = df["dateLastActivity"].dt.year
df_priority["year"] = np.sin(2 * np.pi * df_priority["year"] / (df_priority["year"].max() - df_priority["year"].min()))

df_priority["month"] = df["dateLastActivity"].dt.month
df_priority["month"] = np.sin(2 * np.pi * df_priority["month"] / 12)

df_priority["day"] = df["dateLastActivity"].dt.day

df_priority["day_of_week"] = df["dateLastActivity"].dt.dayofweek
df_priority["day_of_week"] = np.sin(2 * np.pi * df_priority["day_of_week"] / 7)

df_priority["day_of_year"] = df["dateLastActivity"].dt.dayofyear
df_priority["day_of_year"] = np.sin(2 * np.pi * df_priority["day_of_year"] / 365)

df_priority["hour"] = df["dateLastActivity"].dt.hour
df_priority["hour"] = np.sin(2 * np.pi * df_priority["hour"] / 24)

df_priority["is_weekend"] = (df["dateLastActivity"].dt.dayofweek >= 5).astype(int)

df_priority
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(df_priority, df["priority"], test_size=0.2, random_state=42)

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

rf_grid = GridSearchCV(rf_model, rf_params, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
rf_grid.fit(X_train_p, y_train_p)

best_rf_model = rf_grid.best_estimator_
print(f"Best Random Forest Regressor Parameters: {rf_grid.best_params_}")

# Evaluate the best model
y_pred_p = best_rf_model.predict(X_test_p)

print(f"Mean Absolute Error (Priority): {mean_absolute_error(y_test_p, y_pred_p):.2f}")
print(f"Mean Squared Error (Priority): {mean_squared_error(y_test_p, y_pred_p):.2f}")
print(f"R-squared (Priority): {r2_score(y_test_p, y_pred_p):.2f}")

y_pred_p_rounded = np.round(y_pred_p).astype(int)
y_pred_p_rounded = np.clip(y_pred_p_rounded, 1, 5)

print("\nPriority Prediction Classification Report (Rounded Predictions):\n", classification_report(y_test_p, y_pred_p_rounded, zero_division=0))
```
